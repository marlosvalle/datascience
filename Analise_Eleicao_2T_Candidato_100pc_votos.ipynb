```python
# Importando Biblioteca Pandas
import pandas as pd
```

```python
# Cria DataSet importando a base de dados da votação do candidato por Seção Eleitoral
data = pd.read_csv('/content/drive/MyDrive/base_scripts/votacao_secao_2022_BR.csv', sep = ';', encoding='latin-1')
```

```python
# Limite o DataSet somente com os campos necessários para a análise: verificar locais em que determinado candidato teve 100% dos votos e suas características
data = data.drop(['DT_GERACAO','HH_GERACAO','ANO_ELEICAO','CD_TIPO_ELEICAO','CD_ELEICAO','DS_ELEICAO','DT_ELEICAO','TP_ABRANGENCIA','SG_UE','NM_UE','CD_MUNICIPIO','DS_CARGO','NM_VOTAVEL'], axis =1)
```

```python
# Seleciona somente os dados do Segundo Turno da Eleição de 2022 em um novo dataset
data2t = data.query('NR_TURNO == 2')
```

```python
# Coloca o campo do número do candidato como índice
data2t.set_index('NR_VOTAVEL',inplace = True)
```

```python
# Elimina as linhas referente a votos em branco e nulo - codigos 95 e 96, deixando apenas as linhas referente a votação do candidato 13 ou 22
data2t.drop([95,96], inplace = True)
```

```python
# Retira o índice para continuar a análise
data2t = data2t.reset_index()
```

```python
# Cria um novo dataset apenas com os dados agrupados e apenas com registros dos candidatos que receberam votos
data_res = data2t.groupby (['SG_UF','NR_ZONA','NR_SECAO','NM_MUNICIPIO','NM_LOCAL_VOTACAO','DS_LOCAL_VOTACAO_ENDERECO']).sum('NR_VOTAVEL')
```

```python
# Apresenta o resultado do agrupamento - onde NR_VOTAVEL for 35 os dois candidatos receberam votos, diferente disso, apenas um candidato recebeu voto
data_res
```

```python
# Apresenta os dados das urnas - seções - onde o candidato 13 teve 100% dos votos válidos
data_res.query('NR_VOTAVEL == 13')
```

```python
# Apresenta os dados das urnas - seções - onde o candidato 22 teve 100% dos votos válidos
data_res.query('NR_VOTAVEL == 22')
```

```python
# Exporta Arquivo para .csv
data_res.to_csv('candidatos_por_secao.csv')
```
