```python
# Importando Biblioteca Pandas
import pandas as pd
import numpy as np
```

```python
# Importando dados dos Arquivos CSV (Mês 01/2022 a 04/2022, Remuneração Servidores do DF disponível no site http://www.transparencia.df.gov.br/#/downloads#downloadServidores)
dataset1 = pd.read_csv('/content/drive/MyDrive/DataSets/remuneracao_servidores/DF/Remuneracao_2022_01.csv', sep = ';', encoding='latin-1')
dataset2 = pd.read_csv('/content/drive/MyDrive/DataSets/remuneracao_servidores/DF/Remuneracao_2022_02.csv', sep = ';', encoding='latin-1')
dataset3 = pd.read_csv('/content/drive/MyDrive/DataSets/remuneracao_servidores/DF/Remuneracao_2022_03.csv', sep = ';', encoding='latin-1')
dataset4 = pd.read_csv('/content/drive/MyDrive/DataSets/remuneracao_servidores/DF/Remuneracao_2022_04.csv', sep = ';', encoding='latin-1')
```

```python
#Concatena Dataframe, axis = 0 por linha como se fosse append e axis = 1 por coluna; deleta as variáveis Dataframes de apoio
df = pd.concat([dataset1, dataset2, dataset3, dataset4], axis = 0)
del dataset1, dataset2, dataset3, dataset4
```

```python
#Função .info apresenta as informações gerais do dataframe, número linhas, colunas, tipo de dados dos campos, memória alocada
df.info()
```

```python
#Substitui , por . e transforma em float. Só se faz uma vez, na segunda dá erro, pois o dado já foi transformado
df['BRUTO'] = df['BRUTO'].str.replace(',','.').astype(float)
df['LÍQUIDO'] = df['LÍQUIDO'].str.replace(',','.').astype(float)
```

```python
#Análise exploratória usando describe() somente nos campos números do Dataframe
df.describe(include=[np.number])
```

```python
# Oculta as colunas listadas e forma um novo dataset com apenas parte das colunas, como se fosse uma VIEW do dataset principal
df_limited = df.drop(['ANO','MÊS','DESCONTOS A MAIOR','LICENÇA PRÊMIO','IRRF','SEG. SOCIAL','TETO REDUTOR','OUTROS RECEBIMENTOS','BENEFÍCIOS','FUNÇÃO','COMISSÃO CONSELHEIRO','HORA EXTRA','OUTROS DESCONTOS OBRIGATÓRIOS','PAGAMENTOS A MAIOR','VERBAS JUDICIAIS','VALOR DA FUNÇÃO','MATRÍCULA','SITUAÇÃO','VERBAS EVENTUAIS','NOME'], axis = 1)
```

```python
#Apura a média das remunerações bruta e líquida por Cargo/Órgão
df_limited.groupby(['CARGO','CÓDIGO DO ÓRGÃO']).mean()
```

```python
# Filtra Salário Líquido > 80000 no DataFrame Original
df.query('LÍQUIDO > 80000')
```

```python
#Análise geral do dataset
df_limited.describe()
```

```python
# Faz agrupamento usando a função groupby, somando salário líquido
df_gpr = df_limited.groupby(["CPF","CARGO"]).sum("LÍQUIDO")
```

```python
#Filtra pela parcial do CPF informado
cpf = input ("Digite a parcial do CPF ***XXXXXX**: ")
df_gpr_f = df_gpr.query(f"CPF == '***{cpf}**'")
df_gpr_f
```
